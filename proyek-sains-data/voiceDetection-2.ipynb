```python
import os
import pandas as pd
import tsfel
import librosa
```

```python
base_path = "data/voice"

cfg = tsfel.get_features_by_domain()

all_features = []

for person in os.listdir(base_path):
    person_path = os.path.join(base_path, person)
    if not os.path.isdir(person_path):
        continue

    for action in os.listdir(person_path):
        action_path = os.path.join(person_path, action)
        if not os.path.isdir(action_path):
            continue

        for filename in os.listdir(action_path):
            if filename.endswith(".mp3") or filename.endswith(".wav"):
                file_path = os.path.join(action_path, filename)

                try:
                    signal, sr = librosa.load(file_path, sr=None)
                    features = tsfel.time_series_features_extractor(cfg, signal, verbose=0)

                    features['label_orang'] = person
                    features['label_aksi'] = action
                    features['file'] = filename

                    all_features.append(features)

                    print(f"✅ Selesai ekstraksi: {file_path}")
                except Exception as e:
                    print(f"❌ Gagal memproses {file_path}: {e}")

final_df = pd.concat(all_features, ignore_index=True)

final_df.to_csv("data/features_tsfel_all_with_person_label.csv", index=False)
print("✅ Semua fitur berhasil disimpan ke 'data/features_tsfel_all_with_person_label.csv'")
```

```python
data = "data/features_tsfel_all_with_person_label.csv"

df = pd.read_csv(data)

print("✅ Data berhasil dibaca!")
print("Jumlah sample:", len(df))
print("Jumlah kolom :", int(len(df.columns)) - 3)
```

```python
df.head()
```

```python
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
X = df.drop(columns=["label_aksi", "label_orang", "file"])
y_action = df["label_aksi"]
y_person = df["label_orang"]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_action = LabelEncoder()
label_encoder_person = LabelEncoder()

y_action_encoded = label_encoder_action.fit_transform(y_action)
y_person_encoded = label_encoder_person.fit_transform(y_person)

print(label_encoder_action.classes_)
print(label_encoder_person.classes_)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(
    X_scaled, y_action_encoded, test_size=0.2, random_state=42, stratify=y_action_encoded
)

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_scaled, y_person_encoded, test_size=0.2, stratify=y_person_encoded
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

#### Menggunakan semua fitur

```python
model_action = RandomForestClassifier(n_estimators=200, random_state=42)
model_action.fit(X_train_a, y_train_a)
y_pred_action = model_action.predict(X_test_a)

print("\n===== Model: LABEL_AKSI =====")
print(f"Akurasi: {accuracy_score(y_test_a, y_pred_action):.4f}")
print(confusion_matrix(y_test_a, y_pred_action))
print(classification_report(y_test_a, y_pred_action, target_names=label_encoder_action.classes_))
```

```python
model_person = RandomForestClassifier(n_estimators=200, random_state=42)
model_person.fit(X_train_b, y_train_b)
y_pred_action = model_person.predict(X_test_b)

print("\n===== Model: LABEL_ORANG =====")
print(f"Akurasi: {accuracy_score(y_test_b, y_pred_action):.4f}")
print(confusion_matrix(y_test_b, y_pred_action))
print(classification_report(y_test_b, y_pred_action, target_names=label_encoder_person.classes_))
```

#### Rank fitur menggunakan information gain

```python
from sklearn.feature_selection import mutual_info_classif
import matplotlib.pyplot as plt
import seaborn as sns

mutual_information_action = mutual_info_classif(X, y_action_encoded, random_state=42)
mutual_information_person = mutual_info_classif(X, y_person_encoded, random_state=42)
```

```python
mutual_information_df_action = pd.DataFrame({
    "Feature": X.columns,
    "MI_Score": mutual_information_action
}).sort_values(by="MI_Score", ascending=False)

print("🔥 Top 10 Fitur Paling Informatif untuk LABEL_AKSI (buka/tutup):")
print(mutual_information_df_action.head(10), "\n")
```

```python
mutual_information_df_person = pd.DataFrame({
    "Feature": X.columns,
    "MI_Score": mutual_information_person
}).sort_values(by="MI_Score", ascending=False)

print("🔥 Top 10 Fitur Paling Informatif untuk LABEL_ORANG (imam/mufid):")
print(mutual_information_df_person.head(10), "\n")
```

```python
top_n = 20
top_aksi = mutual_information_df_action.sort_values(by="MI_Score", ascending=False).head(top_n)
top_orang = mutual_information_df_person.sort_values(by="MI_Score", ascending=False).head(top_n)

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

sns.barplot(
    data=top_aksi,
    y="Feature",
    x="MI_Score",
    palette="Blues_r",
    ax=axes[0]
)
axes[0].set_title(f"📊 Top {top_n} Fitur berdasarkan Information Gain - LABEL_AKSI", fontsize=13)
axes[0].set_xlabel("Mutual Information Score", fontsize=11)
axes[0].set_ylabel("")
axes[0].tick_params(axis='y', labelsize=9)

sns.barplot(
    data=top_orang,
    y="Feature",
    x="MI_Score",
    palette="Reds_r",
    ax=axes[1]
)
axes[1].set_title(f"📊 Top {top_n} Fitur berdasarkan Information Gain - LABEL_ORANG", fontsize=13)
axes[1].set_xlabel("Mutual Information Score", fontsize=11)
axes[1].set_ylabel("")
axes[1].tick_params(axis='y', labelsize=9)

plt.tight_layout()
plt.show()
```

#### Feature selection

```python
threshold = 0.15

selected_features_action = mutual_information_df_action[mutual_information_df_action['MI_Score'] > threshold]['Feature'].tolist()
selected_features_person = mutual_information_df_person[mutual_information_df_person['MI_Score'] > threshold]['Feature'].tolist()

selected_features_action = list(selected_features_action)
selected_features_person = list(selected_features_person)

X_selected_action = X[selected_features_action]
X_selected_person = X[selected_features_person]

print(f"Jumlah fitur untuk label aksi: {len(selected_features_action)}")
print(f"Jumlah fitur untuk label orang: {len(selected_features_person)}")
```

#### Modeling again, after feature selection

```python
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(
    X_selected_action, y_action, test_size=0.2, random_state=42, stratify=y_action
)

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_selected_person, y_person, test_size=0.2, random_state=42, stratify=y_person
)
```

```python

scaler_action = StandardScaler()
scaler_person = StandardScaler()

X_train_a_scaled = scaler_action.fit_transform(X_train_a)
X_test_a_scaled = scaler_action.transform(X_test_a)

X_train_b_scaled = scaler_person.fit_transform(X_train_b)
X_test_b_scaled = scaler_person.transform(X_test_b)
```

```python
model_action = RandomForestClassifier(n_estimators=200, random_state=42)
model_action.fit(X_train_a_scaled, y_train_a)

y_pred_action = model_action.predict(X_test_a_scaled)

print("=== Evaluasi Model untuk LABEL_AKSI ===")
print(classification_report(y_test_a, y_pred_action))
print("Akurasi:", accuracy_score(y_test_a, y_pred_action))

plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test_a, y_pred_action), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - LABEL_AKSI")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
model_person = RandomForestClassifier(n_estimators=200, random_state=42)
model_person.fit(X_train_b_scaled, y_train_b)

y_pred_person = model_person.predict(X_test_b_scaled)

print("\n=== Evaluasi Model untuk LABEL_ORANG ===")
print(classification_report(y_test_b, y_pred_person))
print("Akurasi:", accuracy_score(y_test_b, y_pred_person))

plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test_b, y_pred_person), annot=True, fmt='d', cmap='OrRd')
plt.title("Confusion Matrix - LABEL_ORANG")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
